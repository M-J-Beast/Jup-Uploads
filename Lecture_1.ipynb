The color of the traces can be specified with passing c="color_name" to either plot or scatter functions. It is able to recognize most of the common color names, as well as hexadesimal code for all of the numbers.

Now let's take a look at tables. In astronomy, tables are often stored in .fits binary format, although astropy.table supports many other formats as well. Let's read a fits table from a server, and take a look at what it has

```python
t=Table.read('https://cdsarc.cds.unistra.fr/ftp/J/A+A/618/A93/table1.fits')
t.info()
print()
```

If we want to see the data in the table, we just need to print it out

```python
t
```

A table consists of named columns that span a certain number of rows. The columns can have different types (e.g., some might contain text, some numbers). Sometimes columns can have units attached to it in the metadata, sometimes they can have comments, both for the table as a whole and for the individual columns.

To access a column we just need to refer to index it using its name. Keep in mind that capitalization in the column names (and in Python as a whole) does matter, e.g., x != X

```python
print(t['X'])
print()

print(t['X'].data) #
print()

xyz = t[['X', 'Y', 'Z']] #
print(xyz)
```

We can also call on an individual row by its numerical index

```python
print(xyz[0:10]) #

print()
print(xyz['X'][1]) #
```

In general, we can treat tables as ordered groups of arrays - e.g., if we want to access data on a certain source, all of those data would be stored in the same row, and the same index would work on all of the columns.

While it is possible to add new rows to the table, it is not a basic task. Nor should it be, as in this case you need to be creating new data that has to be consistent with existing data. On the other hand, we can easily add a new column

```python
xyz['new_column1']=30 #
xyz['new_column2']=np.arange(len(xyz)) #
xyz['new_column3']=xyz['X']*2 #

print(xyz)
```

Create a column that would be a hypotenuse in 3d of X, Y, and Z.

```python
#put your code here
xyz['new_column3D']= np.sqrt(xyz['X']**2 + xyz['Y']**2 + xyz['Z']**2)

print(xyz)
```
